```python
import os
while True:
    print("""""  Welcome to Docker Commands Console   """"")
    print("""
    
    Enter your choice 
    
    1-Start Docker
    2-Check Docker Status
    3-Pull an image
    4-Check images
    5-Removing a image
    6-Run a container 
    7-Check all running Containers
    8-Remove a Container
    9-Go to Main menu
    """)
    x=int(input(" Enter your choice "))
    if x==1:
        os.system("systemctl start docker")
        print("Docker Started")
        input("Enter to continue")
    elif x==2:
        os.system("systemctl status docker")
        print("It will show whether docker is active or not ")
        input("Enter to continue")
    elif x==3:
        image_name=input("Enter image name that you want to download ")
        os.system("docker pull {}".format(image_name))
        input("Enter to continue")
    elif x==4:
        print("All the images in docker will be shown.")
        os.system("docker images")
        input("Enter to continue")
    elif x==5:
        image_name=input("Enter image name that you want to download ")
        os.system("docker image rm {}".format(image_name))
        input("Enter to continue")
    elif x==6:
        container_name=input("Enter name of container that you want to run")
        os.system("docker start {}".format(container_name))
        input("Enter to continue")
    elif x==7:
        print("All the containers running will be shown.")
        os.system("docker ps -a")
        input("Enter to continue")
    elif x==8:
        container_name=input("Enter name of container that you want to remove.")
        os.system("docker rm {}".format(container_name))
        input("Enter to continue")
    elif x==9:    
        break
    else:
        print("Invalid choice")
        input("Enter to Continue")
        continue
    os.system("clear")    
```
